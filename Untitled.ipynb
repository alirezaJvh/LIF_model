```python
from LIF.LIF import LIF
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
parameters = {
    'C': 1,
    'R': 0.12,
    'u_rest': 3,
    'I': 0.5,
    'threshold': 3.05,
    'start_time': 10,
    'end_time': 150,
    'dt': 0.1
}
```

```python
model = LIF(parameters)
sim= model.simulate()
```

```python
len(sim['u'])
```

```python
df = pd.DataFrame(sim)
```

```python
plt.figure(figsize=(15,5))
plt.plot(sim['time'], sim['u'])
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(sim['time'], model.get_I_history())
plt.show()
```

```python
i = np.arange(2, 20, 1)
T = model.interval(i)
F = 1 / T
```

```python
# plt.figure(figsize=(5,5))
plt.axis('equal')
plt.plot(i, F)
plt.show()
```
